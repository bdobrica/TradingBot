```python
import pandas as pd

df = pd.DataFrame(columns = ['price', 'symbol', 'time', 'volume'])
```

```python
import json
import datetime

def on_message(ws, message):
    message = json.loads(message)
    if 'data' not in message:
        return
    if 'type' not in message:
        return
    if message['type'] != 'trade':
        return
    data = message['data']
    if not isinstance(data, list):
        return
    
    for item in data:
        if not isinstance(item, dict):
            continue
        if 'p' not in item or 's' not in item or 't' not in item or 'v' not in item:
            continue
        
        global df
        df = df.append({
            'price': item['p'],
            'symbol': item['s'],
            'time': datetime.datetime.fromtimestamp(item['t'] / 1000.0).strftime('%Y-%m-%d %H:%M:%S.%f'),
            'volume': item['v']
        }, ignore_index = True)
```

```python
if __name__ == "__main__":
    websocket.enableTrace(True)
    ws = websocket.WebSocketApp("wss://ws.finnhub.io?token=btbpfin48v6p15lfhoig",
        on_message = on_message,
        on_error = on_error,
        on_close = on_close)
    ws.on_open = on_open
    ws.run_forever()
```
